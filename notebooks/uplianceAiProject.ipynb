```python
from sklearn.linear_model import LinearRegression
```

```python
ages = np.array([42, 38, 28, 27, 30, 29, 35, 33, 31, 25]).reshape(-1, 1)
```

```python
orders = np.array([15, 14, 12, 10, 9, 8, 8, 6, 5, 7])
```

```python
lr1 = LinearRegression()
```

```python
lr1.fit(ages, orders)
```

```python
import matplotlib.pyplot as plt
```

```python
ages_range = np.linspace(min(ages), max(ages), 100).reshape(-1, 1)
predicted_orders = lr1.predict(ages_range)

# Plot the data points
plt.figure(figsize=(10, 6))
plt.scatter(ages, orders, color='blue', label='Actual Data', s=100)
plt.plot(ages_range, predicted_orders, color='red', label='Regression Line', linewidth=2)

# Add labels, title, and legend
plt.title('Linear Regression: Total Orders by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Total Orders', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
